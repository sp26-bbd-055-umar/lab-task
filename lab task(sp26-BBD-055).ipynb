```python
# 1.
num = 1

while num <= 10:
    print(num)
    num += 1
```

```python
# 2.
num = int(input("Enter a starting number: "))
while num >= 0:
    print(num)
    num -= 1
```

```python
# 3.
n = int(input("Enter a positive number: "))
total = 0
i = 1
while i <= n:
    total += i
    i += 1
print("Sum =", total)
```

```python
# 4.
num = int(input("Enter a number: "))
i = 1
while i <= 10:
    print(num, "x", i, "=", num * i)
    i += 1
```

```python
# 5.
num = int(input("Enter a positive number: "))
while num <= 0:
    print("Invalid input!")
    num = int(input("Enter a positive number: "))
print("Valid number entered:", num)
```

```python
# 6.
password = "admin123"
user_input = input("Enter password: ")
while user_input != password:
    print("Wrong password!")
    user_input = input("Try again: ")
print("Access granted!")
```

```python
# 7.
num = int(input("Enter a number: "))
count = 0
while num != 0:
    num //= 10
    count += 1
print("Number of digits:", count)
```

```python
# 8.
total = 0
num = int(input("Enter a number (0 to stop): "))
while num != 0:
    total += num
    num = int(input("Enter a number (0 to stop): "))
print("Final total =", total)
```

```python
# 9.
num = 101
while True:
    if num % 7 == 0:
        print("First multiple of 7 > 100 is:", num)
        break
    num += 1
```

```python
secret = 17
guess = int(input("Guess the number: "))
attempts = 1
while guess != secret:
    if guess > secret:
        print("Too high!")
    else:
        print("Too low!")
    guess = int(input("Try again: "))
    attempts += 1
print("Correct! Attempts:", attempts)
```
